### 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75 # control size of points - 'X's

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             color='green', label='Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.6 Adding mathematical functions

### 1.7 We need another axis!

```python
x = np.linspace(0, 6, 100) # (start no, stop no, no. of numbers)
# creates an array of evenly spaced numbers over a specified interval.
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

### 1.8 Saving to disc

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
```

```python
plt.style.use('fivethirtyeight')
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
# y-value, starting x, ending x
```

```python
plt.style.use('default')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
```

## Exercise 2 :  A simple mathematical plot

```python
x = np.linspace(0, 50, 300)
y = np.exp(-x/10) * np.sin(x)

plt.plot(x, y)
plt.grid(alpha=0.25)
plt.xlabel('x-values', color='darkblue', fontsize='15')
plt.ylabel('y-values', color='darkblue', fontsize='15')
plt.title('$y=e^{(-x/10)}sin(x)$', fontsize='20')
plt.ylim([-1,1])

plt.show()
```
